```python
import os
```

```python
hq_mask_folder_path = '/home/miguel/GI/1.5 - Synthetic Data Generation/diffuse-gen/diffuse-gen/guided_diffusion/segmented-images-roberta-hq-masks'
```

```python
for cv_subject in os.listdir(hq_mask_folder_path):
    cv_subject_path = os.path.join(hq_mask_folder_path, cv_subject, 'masked-images')
    if not os.path.isdir(cv_subject_path):
        continue
    for img_file in os.listdir(cv_subject_path):
        img_file_path = os.path.join(cv_subject_path, img_file)
        if 'synthetic' in img_file:
            os.remove(img_file_path)
            print(f'Removed {img_file_path}')
```
